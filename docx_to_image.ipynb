```python
from docx import Document
from io import BytesIO
from PIL import Image
from IPython.display import display
```

```python
# Load the DOCX file
doc_path = "/content/ban_5.docx"  # Replace with your DOCX file
doc = Document(doc_path)
```

```python

# Extract images
image_count = 0
for rel in doc.part.rels:
    if "image" in doc.part.rels[rel].target_ref:
        image_count += 1
        image_data = doc.part.rels[rel].target_part.blob  # Extract image bytes
        img = Image.open(BytesIO(image_data))  # Open as PIL Image
        img_format = img.format if img.format else "PNG"  # Detect format
        img.save(f"extracted_image_{image_count}.{img_format.lower()}")  # Save image
        print(f"Saved: extracted_image_{image_count}.{img_format.lower()}")

        # Display the extracted image
        display(img)  # This line is added to display the image


if image_count == 0:
    print("No images found in the DOCX file.")
```
